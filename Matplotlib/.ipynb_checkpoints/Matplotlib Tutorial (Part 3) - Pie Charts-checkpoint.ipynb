```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [60, 40, 50, 70]
labels = ['Sixty', 'Forty', 'Fifty', 'Seventy']
#colors = ['Blue', 'red', 'Yellow', 'green']
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']


plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

**Shadow to our Chart**

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]


plt.pie(slices, labels=labels, explode=explode, shadow=True, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

**Strat Angle**

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]


plt.pie(slices, labels=labels,
        explode=explode,
        shadow=True,
        startangle=90,
        wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

**Display percentage**

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]


plt.pie(slices, labels=labels,
        explode=explode,
        shadow=True, # Shadow to our Chart
        startangle=90, #Strat Angle
        autopct="%1.1f%%", # format string to specify the pecentage
        wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
