```python
import pandas as pd
import os

from matplotlib import pyplot as plt

os.chdir('/Users/Maanvitha/Documents/File-Management/matplotlib/') # Change Directory

plt.style.use('seaborn')

data = pd.read_csv('data7.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, 
            c=ratio,
            #s=sizes, # S for set the size the scatters
            cmap = 'summer',
            edgecolor='black',
            linewidth=1,
            alpha=0.75
           )

plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
